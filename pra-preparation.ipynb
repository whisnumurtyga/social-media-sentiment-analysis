# Merger Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        print(f'Folder: {root}')
        for file in files:
            print(f'File: {os.path.join(root, file)}')
```

```python
# Ganti path sesuai dengan direktori yang ingin kamu telusuri
folder_path = 'raw-data'

# Inisialisasi DataFrame kosong
all_data = pd.DataFrame()

# Melakukan iterasi untuk setiap file di dalam direktori dan subdirektori
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.csv'):  # Memastikan hanya file CSV yang digunakan
            file_path = os.path.join(root, file)
            data = pd.read_csv(file_path)
            all_data = pd.concat([all_data, data], ignore_index=True)
```

```python
all_data_in['location'] = all_data_in['location'].str.lower()
all_data_in['location'] = all_data_in['location'].astype('str')
all_data_in.head()
```

```python
# Split kolom 'location', tambahkan kolom baru 'loc1' dan 'loc2'
split_values = all_data_in['location'].str.split(', ', expand=True)
num_columns = len(split_values.columns)

# Tambahkan kolom 'loc1' dan 'loc2' dengan menangani hasil split yang menghasilkan string kosong
for i in range(num_columns):
    col_name = f'loc{i+1}'
    all_data_in[col_name] = split_values[i].fillna('')

all_data_in.head()
```
