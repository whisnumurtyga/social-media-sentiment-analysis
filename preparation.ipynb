```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/data-jatim.csv')
df.head()
```

```python
text_df = df.iloc[:,[2]]
text_df.head()
# text_df.to_csv('dataset/text_df.csv', index=False)
```

```python
import string
import re  # Modul untuk ekspresi reguler
```

```python
def cleaning_text(kalimat):
    kalimat = re.sub(r'@\w+', '', kalimat)  # Menghapus mention (username Twitter)
    kalimat = re.sub(r'https?://\S+', '', kalimat)  # Menghapus URL
    kalimat = re.sub(r'\t|\n|\r', ' ', kalimat)  # Mengganti karakter tab, baris baru, dan carriage return dengan spasi
    kalimat = kalimat.encode('ascii', 'ignore').decode('ascii')  # Menghapus karakter non-ASCII
    kalimat = re.sub(r'@\w+|https?://\S+|#\w+', '', kalimat)  # Menghapus mention, URL, dan hashtag
    kalimat = re.sub(r'www\.\S+', '', kalimat)  # Menghapus URL dengan format tertentu
    kalimat = re.sub(r'\d+', '', kalimat)  # Menghapus angka
    kalimat = kalimat.translate(str.maketrans("", "", string.punctuation))  # Menghapus tanda baca
    kalimat = re.sub(r'"', '"', kalimat)  # Menghapus karakter quote yang tidak standar
    kalimat = re.sub(r'[^a-zA-Z0-9\s]', '', kalimat)  # Menghapus karakter selain huruf, angka, dan spasi
    kalimat = kalimat.replace('\n', ' ')  # Mengganti karakter baris baru dengan spasi
    kalimat = kalimat.lower()  # Mengubah teks menjadi lowercase
    kalimat = ' '.join([word for word in kalimat.split() if len(word) > 1])  # Menghilangkan kata-kata yang hanya terdiri dari satu karakter
    kalimat = ' '.join(re.findall(r'\b\w+\b', kalimat))  # Menjaga hanya kata-kata dan menghapus karakter-karakter spesial lainnya
    kalimat = kalimat.strip()  # Menghapus spasi ekstra di awal dan akhir teks
    kalimat = re.sub(r'\s+', ' ', kalimat)  # Mengganti beberapa spasi berturut-turut dengan satu spasi

    return kalimat
```

```python
text_df['text_clean'] = text_df['full_text'].apply(cleaning_text)
text_df.head()
```

```python
text_df.to_csv('dataset/text-clean.csv', index=False)
```
